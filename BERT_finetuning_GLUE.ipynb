```
# Tokenize all of the sentences and map the tokens to their word IDs.

# dictionary of lists
import collections
input_ids = collections.defaultdict(list)
attention_masks = collections.defaultdict(list)
# define sequence lengths
seq_len = list(range(0,512,64))

# for every dataset
for i in data_names:
  # For every sentence...
  for j in range(0, len(sent_quest_1[i])):
    # `encode_plus` will:
    #   (1) Tokenize the sentence.
    #   (2) Prepend the `[CLS]` token to the start.
    #   (3) Append the `[SEP]` token to the end.
    #   (4) Map tokens to their IDs.
    #   (5) Pad or truncate the sentence to `max_length`
    #   (6) Create attention masks for [PAD] tokens.

    # if a sentence/question pair exists, tokenize sentence/question pair
    if i in sent_quest_2.keys():
      # Different sequence lengths, depending on maximal sequence length of 
      # respective dataset (steps of 64)
      for count in range(1, len(seq_len)):
        if seq_len[count-1] < max_len[i] <= seq_len[count]:
          encoded_dict = tokenizer.encode_plus(
                        str(sent_quest_1[i][j]),     # Sentence to encode.
                        str(sent_quest_2[i][j]), 
                        add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                        max_length = seq_len[count],           # Pad & truncate all sentences.   (max_length = 128, just in case)
                        pad_to_max_length = True,
                        return_attention_mask = True,   # Construct attn. masks.
                        return_tensors = 'pt',     # Return pytorch tensors.
                        )
      # Add the encoded sentence to the list.    
      input_ids[i].append(encoded_dict['input_ids'])
    
      # And its attention mask (simply differentiates padding from non-padding).
      attention_masks[i].append(encoded_dict['attention_mask'])

    elif i not in sent_quest_2.keys():
      for count in range(1, len(seq_len)):
        if max_len[i] <= seq_len[count]:
          encoded_dict = tokenizer.encode_plus(
                        str(sent_quest_1[i][j]),                      # Sentence to encode. 
                        add_special_tokens = True, # Add '[CLS]' and '[SEP]'
                        max_length = seq_len[count],           # Pad & truncate all sentences.   (max_length = 128, just in case)
                        pad_to_max_length = True,
                        return_attention_mask = True,   # Construct attn. masks.
                        return_tensors = 'pt',     # Return pytorch tensors.
                   )
    
      # Add the encoded sentence to the list.    
      input_ids[i].append(encoded_dict['input_ids'])
    
      # And its attention mask (simply differentiates padding from non-padding).
      attention_masks[i].append(encoded_dict['attention_mask'])

# Convert the lists into tensors.
for i in data_names:
  input_ids[i] = torch.cat(input_ids[i], dim=0)    # Concatenates the given sequence of seq tensors in the given dimension
  attention_masks[i] = torch.cat(attention_masks[i], dim=0)
  if i == 'STSB':
    labels[i] = torch.tensor(labels[i], ).type(torch.DoubleTensor)
  else:
    labels[i] = torch.tensor(labels[i], )
```

Training & Validation Split

```
from torch.utils.data import TensorDataset, random_split

dataset = collections.defaultdict(torch.utils.data.dataset.TensorDataset)

print(data_names)
for i in data_names:
  #print(input_ids[i])
  #print(attention_masks[i])
  print(i)
  print(labels[i])

# Combine the training inputs into a TensorDataset.
for i in data_names:
  dataset[i] = TensorDataset(input_ids[i], attention_masks[i], labels[i])

# Create a 90-10 train-validation split.

train_size = {}
val_size = {}
# Calculate the number of samples to include in each set.
for i in data_names:
  train_size[i] = int(0.9 * len(dataset[i]))
  val_size[i] = len(dataset[i]) - train_size[i]

train_dataset = {}
val_dataset = {}

# Divide the dataset by randomly selecting samples.
for i in data_names:
  train_dataset[i], val_dataset[i] = random_split(dataset[i], [train_size[i], val_size[i]])
  
  print('{:>5,} training samples'.format(train_size[i]))
  print('{:>5,} validation samples'.format(val_size[i]))
```

```
# create iterator

from torch.utils.data import DataLoader, RandomSampler, SequentialSampler

# Data loader: combines a dataset and a sampler, and provides an iterable over the given dataset; (iterable: IterationsfÃ¤higkeit)
# DataLoader by default constructs a index sampler that yields integral indices

# The DataLoader needs to know our batch size for training, so we specify it 
# here. For fine-tuning BERT on a specific task, the authors recommend a batch 
# size of 16 or 32.
batchsizes = [16, 32, 48, 64]
# 16, 32 and 48 in BERT paper

train_dataloaders = collections.defaultdict(dict)
validation_dataloaders = collections.defaultdict(dict)
#counter = 0
for i in data_names:
  for bs in batchsizes:
    # Create the DataLoaders for our training and validation sets.
    # We'll take training samples in random order.
    train_dataloaders[i][bs] = DataLoader(
              dataset = train_dataset[i],  # The training samples.
              sampler = RandomSampler(train_dataset[i]), # Select batches randomly
              batch_size = bs # Trains with this batch size.
            )
    # For validation the order doesn't matter, so we'll just read them sequentially.
    validation_dataloaders[i][bs] = DataLoader(
                dataset = val_dataset[i], # The validation samples.
                sampler = SequentialSampler(val_dataset[i]), # Pull out batches sequentially.
                batch_size = bs # Evaluate with this batch size.
            )
    #counter += 1
```

Train Classification Model

```
from transformers import BertForSequenceClassification, AdamW, BertConfig

# Load BertForSequenceClassification, the pretrained BERT model with a single 
# linear classification layer on top. 
model = {}
num_labels = {}

for i in data_names:

  if len(torch.unique(labels[i])) < 10:   # because one task is a regression task 
    
    # number of labels for each dataset
    num_labels[i] = len(torch.unique(labels[i]))

  else:
    # number of labels for each dataset; for regression task: num_labels should be 1
    num_labels[i] = 1
  
  model[i] = BertForSequenceClassification.from_pretrained(
      "bert-base-uncased", # Use the 12-layer BERT model, with an uncased vocab.
      num_labels = num_labels[i], # The number of output labels--2 for binary classification.
      # You can increase this for multi-class tasks.   
      output_attentions = False, # Whether the model returns attentions weights.
      output_hidden_states = False, # Whether the model returns all hidden-states.
      )
  
  # Tell pytorch to run this model on the GPU.
  model[i].cuda()
```

Optimizer & Learning Rate Scheduler

```
# Note: AdamW is a class from the huggingface library (as opposed to pytorch) 
# I believe the 'W' stands for 'Weight Decay fix"

learning_rates = [1e-5, 2e-5, 3e-5, 4e-5, 5e-5]
#epsilons = [1e-8, 2e-8, 3e-8]
optimizer = collections.defaultdict(dict)

for i in data_names:
  for lern in learning_rates:
    optimizer[i][lern] = AdamW(model[i].parameters(),
                  lr = 5e-5, # args.learning_rate - default is 5e-5, our notebook had 2e-5
                  eps = 1e-8 # args.adam_epsilon  - default is 1e-8.
                )
```

```
from transformers import get_linear_schedule_with_warmup

# Number of training epochs. The BERT authors recommend between 2 and 4. 
epochs = 2

# Total number of training steps is [number of batches] x [number of epochs]. 
# (Note that this is not the same as the number of training samples).

# For each batch size a different number of steps is applied
total_steps = collections.defaultdict(dict)

for i in data_names:
  for key, value in train_dataloaders[i].items():
    total_steps[i][key] = len(value) * epochs
  

# print number of steps
print(total_steps)

# Create the learning rate scheduler.
schedulers = collections.defaultdict(lambda: collections.defaultdict(lambda: collections.defaultdict(int)))

for i in data_names:
  for lern in learning_rates:
    for step in total_steps[i].items():
      schedulers[i][lern][step] = get_linear_schedule_with_warmup(optimizer[i][lern], 
                        num_warmup_steps = 0, # Default value in run_glue.py
                        num_training_steps = step[1])
```

```
# We'll store a number of quantities such as training and validation loss, 
# validation accuracy, and timings.
training_stats = collections.defaultdict(dict)

for i in data_names:
  for bs in batchsizes:
    for lern in learning_rates:
      finetuning(train_dataloaders[i][bs], validation_dataloaders[i][bs], schedulers[i][bs][lern], training_stats[i], model[i], optimizer[i][lern])
```
